```python
import sys
# Add the directory containing the oqs Python module to the system path
sys.path.append('/content/liboqs/python')

import oqs

# Choose an ML-DSA algorithm, e.g., ML-DSA-44 (NIST security level 2)
sig_alg = "ML-DSA-44"
print(f"Using ML-DSA algorithm: {sig_alg}")

# Ensure the algorithm is supported
if oqs.is_sig_supported(sig_alg):
    # Instantiate the signature scheme
    sig = oqs.Signature(sig_alg)

    # 1. Key Generation: Create a public and private key pair
    print("Generating key pair...")
    public_key = sig.generate_keypair()
    print("Key pair generated.")

    # 2. Signing: Create a test message and sign it with the private key
    message = b"Test message for ML-DSA signature."
    print("\nOriginal message:", message)
    signature = sig.sign(message)
    print("Message signed.")

    # 3. Verification: Verify the signature using the public key
    print("Verifying signature...")
    is_verified = sig.verify(message, signature, public_key)

    # Print the result
    if is_verified:
        print("\nVerification successful! The signature is valid. ✅")
    else:
        print("\nVerification failed! The signature is invalid. ❌")
else:
    print(f"Error: The algorithm '{sig_alg}' is not supported by liboqs.")
```

```python
# Clean up previous builds to start fresh
rm -rf /content/liboqs

# Install build dependencies
apt-get update -y > /dev/null
apt-get install -y cmake build-essential > /dev/null

# Clone the liboqs repository
git clone https://github.com/open-quantum-safe/liboqs.git > /dev/null

# Change into the repository directory
cd /content/liboqs

# Configure and build the C library. We use the `-DOQS_USE_NIST_KAT` flag to ensure ML-DSA is built.
cmake -DBUILD_SHARED_LIBS=ON -DOQS_BUILD_ONLY_LIB=OFF -DOQS_USE_NIST_KAT=ON . > /dev/null
make > /dev/null

# Verify that the library was built
echo "liboqs library built successfully."

# Write the C code to a file named 'test_mldsa.c'
%%writefile /content/liboqs/test_mldsa.c
#include <stdio.h>
#include <string.h>
#include <oqs/oqs.h>

int main() {
    printf("Starting ML-DSA-44 test...\n");

    // The ML-DSA-44 algorithm is NIST Dilithium2
    OQS_SIG *sig = OQS_SIG_new(OQS_SIG_alg_ml_dsa_44);
    if (sig == NULL) {
        printf("Failed to create signature algorithm. Make sure liboqs was built with ML-DSA-44 support.\n");
        return 1;
    }

    uint8_t *public_key = (uint8_t *)malloc(sig->length_public_key);
    uint8_t *secret_key = (uint8_t *)malloc(sig->length_secret_key);

    // Key Generation
    OQS_STATUS status = OQS_SIG_keypair(sig, public_key, secret_key);
    if (status != OQS_SUCCESS) {
        printf("Keypair generation failed!\n");
        free(public_key);
        free(secret_key);
        OQS_SIG_free(sig);
        return 1;
    }
    printf("Keypair generated successfully.\n");

    // Message and Signature
    const uint8_t *message = (const uint8_t *)"This is a test message for ML-DSA.";
    size_t message_len = strlen((const char *)message);
    uint8_t *signature = (uint8_t *)malloc(sig->length_signature);
    size_t signature_len;

    // Signing the message
    status = OQS_SIG_sign(sig, signature, &signature_len, message, message_len, secret_key);
    if (status != OQS_SUCCESS) {
        printf("Signing failed!\n");
        free(public_key);
        free(secret_key);
        free(signature);
        OQS_SIG_free(sig);
        return 1;
    }
    printf("Message signed successfully.\n");

    // Verification
    status = OQS_SIG_verify(sig, message, message_len, signature, signature_len, public_key);
    if (status == OQS_SUCCESS) {
        printf("Verification successful! The signature is valid. ✅\n");
    } else {
        printf("Verification failed! The signature is invalid. ❌\n");
    }

    // Clean up
    free(public_key);
    free(secret_key);
    free(signature);
    OQS_SIG_free(sig);

    return 0;
}
```

## custom-crypto-module.js

This file contains the JavaScript wrapper to load and interact with the WebAssembly module compiled from the `liboqs` C library. It exposes functions for key generation, signing, and verification.

**Note**: This code assumes that `oqs_mldsa.js` and `oqs_mldsa.wasm` are available in the same directory or a location accessible by the Node.js environment.

```python
// custom-crypto-module.js

// Import the WebAssembly module loader generated by Emscripten
// This assumes oqs_mldsa.js is in the same directory
const OQSMldsaModule = require('./oqs_mldsa');

let oqsModule = null;
let mldsaSig = null; // Pointer to the OQS_SIG object in WASM memory
let publicKeyLen = 0;
let secretKeyLen = 0;
let signatureLen = 0;

// Function to initialize the WASM module and get the signature object
async function initializeWasm() {
    if (oqsModule === null) {
        // Instantiate the WASM module. The OQSMldsaModule function
        // is the default export from oqs_mldsa.js
        oqsModule = await OQSMldsaModule();

        // Get the pointer to the ML-DSA-44 OQS_SIG object in WASM
        mldsaSig = oqsModule._get_ml_dsa_44_sig();

        if (mldsaSig === null) {
            throw new Error("Failed to initialize ML-DSA-44 signature algorithm in WASM.");
        }

        // Get the required buffer lengths for keys and signatures
        publicKeyLen = oqsModule._get_public_key_len(mldsaSig);
        secretKeyLen = oqsModule._get_secret_key_len(mldsaSig);
        signatureLen = oqsModule._get_signature_len(mldsaSig);

        if (publicKeyLen === 0 || secretKeyLen === 0 || signatureLen === 0) {
             throw new Error("Failed to get key or signature lengths from WASM.");
        }

        console.log("WASM module initialized successfully.");
        console.log(`Public Key Length: ${publicKeyLen}`);
        console.log(`Secret Key Length: ${secretKeyLen}`);
        console.log(`Signature Length: ${signatureLen}`);

    }
    return oqsModule;
}

// Function to generate a key pair
async function generateKeyPair() {
    await initializeWasm();

    // Allocate memory in the WASM heap for public and secret keys
    const publicKeyPtr = oqsModule._malloc(publicKeyLen);
    const secretKeyPtr = oqsModule._malloc(secretKeyLen);

    if (publicKeyPtr === 0 || secretKeyPtr === 0) {
        // Free allocated memory before throwing error
        if (publicKeyPtr !== 0) oqsModule._free(publicKeyPtr);
        if (secretKeyPtr !== 0) oqsModule._free(secretKeyPtr);
        throw new Error("Failed to allocate memory for keys in WASM heap.");
    }

    try {
        // Call the WASM function to generate the key pair
        const status = oqsModule._generate_keypair_wasm(mldsaSig, publicKeyPtr, secretKeyPtr);

        if (status !== 0) { // Assuming OQS_SUCCESS is 0
            throw new Error(`WASM keypair generation failed with status: ${status}`);
        }

        // Copy the keys from the WASM heap to JavaScript ArrayBuffers
        const publicKey = new Uint8Array(oqsModule.HEAPU8.buffer, publicKeyPtr, publicKeyLen).slice();
        const secretKey = new Uint8Array(oqsModule.HEAPU8.buffer, secretKeyPtr, secretKeyLen).slice();

        return { publicKey: Buffer.from(publicKey), secretKey: Buffer.from(secretKey) };

    } finally {
        // Free the allocated memory in the WASM heap
        if (publicKeyPtr !== 0) oqsModule._free(publicKeyPtr);
        if (secretKeyPtr !== 0) oqsModule._free(secretKeyPtr);
    }
}

// Function to sign a message
async function sign(message, secretKey) {
     await initializeWasm();

     // Ensure secretKey is a Buffer
     const secretKeyBuffer = Buffer.from(secretKey);
     if (secretKeyBuffer.length !== secretKeyLen) {
         throw new Error(`Invalid secret key length. Expected ${secretKeyLen}, got ${secretKeyBuffer.length}`);
     }

     // Allocate memory in the WASM heap for the message, secret key, signature, and signature length
     const messagePtr = oqsModule._malloc(message.length);
     const secretKeyPtr = oqsModule._malloc(secretKeyBuffer.length);
     const signaturePtr = oqsModule._malloc(signatureLen);
     const signatureLenPtr = oqsModule._malloc(8); // size_t is typically 8 bytes on 64-bit systems

     if (messagePtr === 0 || secretKeyPtr === 0 || signaturePtr === 0 || signatureLenPtr === 0) {
         // Free allocated memory before throwing error
         if (messagePtr !== 0) oqsModule._free(messagePtr);
         if (secretKeyPtr !== 0) oqsModule._free(secretKeyPtr);
         if (signaturePtr !== 0) oqsModule._free(signaturePtr);
         if (signatureLenPtr !== 0) oqsModule._free(signatureLenPtr);
         throw new Error("Failed to allocate memory for signing in WASM heap.");
     }

     try {
         // Copy message and secret key to the WASM heap
         oqsModule.HEAPU8.set(message, messagePtr);
         oqsModule.HEAPU8.set(secretKeyBuffer, secretKeyPtr);
         oqsModule.setValue(signatureLenPtr, signatureLen, 'i64'); // Initialize signature length pointer

         // Call the WASM function to sign the message
         const status = oqsModule._sign_wasm(mldsaSig, signaturePtr, signatureLenPtr, messagePtr, message.length, secretKeyPtr);

         if (status !== 0) { // Assuming OQS_SUCCESS is 0
             throw new Error(`WASM signing failed with status: ${status}`);
         }

         // Get the actual signature length after signing
         const actualSignatureLen = oqsModule.getValue(signatureLenPtr, 'i64');

         // Copy the signature from the WASM heap to a JavaScript ArrayBuffer
         const signature = new Uint8Array(oqsModule.HEAPU8.buffer, signaturePtr, actualSignatureLen).slice();

         return Buffer.from(signature);

     } finally {
         // Free the allocated memory in the WASM heap
         if (messagePtr !== 0) oqsModule._free(messagePtr);
         if (secretKeyPtr !== 0) oqsModule._free(secretKeyPtr);
         if (signaturePtr !== 0) oqsModule._free(signaturePtr);
         if (signatureLenPtr !== 0) oqsModule._free(signatureLenPtr);
     }
}

// Function to verify a signature
async function verify(message, signature, publicKey) {
    await initializeWasm();

    // Ensure signature and publicKey are Buffers
    const signatureBuffer = Buffer.from(signature);
    const publicKeyBuffer = Buffer.from(publicKey);

    if (publicKeyBuffer.length !== publicKeyLen) {
        throw new Error(`Invalid public key length. Expected ${publicKeyLen}, got ${publicKeyBuffer.length}`);
    }

    // Allocate memory in the WASM heap for the message, signature, and public key
    const messagePtr = oqsModule._malloc(message.length);
    const signaturePtr = oqsModule._malloc(signatureBuffer.length);
    const publicKeyPtr = oqsModule._malloc(publicKeyBuffer.length);

    if (messagePtr === 0 || signaturePtr === 0 || publicKeyPtr === 0) {
        // Free allocated memory before throwing error
        if (messagePtr !== 0) oqsModule._free(messagePtr);
        if (signaturePtr !== 0) oqsModule._free(signaturePtr);
        if (publicKeyPtr !== 0) oqsModule._free(publicKeyPtr);
        throw new Error("Failed to allocate memory for verification in WASM heap.");
    }

    try {
        // Copy message, signature, and public key to the WASM heap
        oqsModule.HEAPU8.set(message, messagePtr);
        oqsModule.HEAPU8.set(signatureBuffer, signaturePtr);
        oqsModule.HEAPU8.set(publicKeyBuffer, publicKeyPtr);

        // Call the WASM function to verify the signature
        const status = oqsModule._verify_wasm(mldsaSig, messagePtr, message.length, signaturePtr, signatureBuffer.length, publicKeyPtr);

        // Assuming OQS_SUCCESS is 0, verification is successful if status is 0
        return status === 0;

    } finally {
        // Free the allocated memory in the WASM heap
        if (messagePtr !== 0) oqsModule._free(messagePtr);
        if (signaturePtr !== 0) oqsModule._free(signaturePtr);
        if (publicKeyPtr !== 0) oqsModule._free(publicKeyPtr);
    }
}

// Clean up the OQS_SIG object when the module is no longer needed
// In a Cloud Function, this might not be strictly necessary as the environment
// is torn down after execution, but it's good practice.
function cleanup() {
    if (oqsModule && mldsaSig) {
        oqsModule._free_sig_wasm(mldsaSig);
        mldsaSig = null;
        oqsModule = null;
        console.log("WASM module cleaned up.");
    }
}

// Export the functions to be used by index.js
module.exports = {
    generateKeyPair,
    sign,
    verify,
    initializeWasm, // Export initialize for potential eager loading
    cleanup
};
```
